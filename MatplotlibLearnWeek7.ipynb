# 3D Axes

Untuk membuat 3d axes kita hanya perlu menambahkan parameter projection='3d' pada fungsi plt.axes()

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
plt.show()
```

note: agar 3d object bisa kita gerakan (dirotate) kita perlu menggunakan magic command %matplotlib notebook 

# 3D Line Plot

Untuk membuat 3D line plot kita hanya perlu memanggil fungsi ax.plot() dengan menyertakan 3 buah nilai untuk merepresentasikan koordinat x,y, dan z

```python
z = np.linspace(0,1,50) #paramter pertama: nilai awal, paramter kedua nilai akhri, paramter ketiga banyaknya nilai
x = z * np.sin(20*z)
y = z * np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot(x,y,z,color='yellow') #parameter color adalah parameter optional
ax.set(title='Contoh 3D line plot',
       xlabel = 'sumbu X',
       ylabel = 'sumbu Y',
       zlabel = 'sumbu Z')
plt.show()
```

# 3D Scatter Plot


Untuk membuat 3D Scatter plot kita hanya perlu memanggil fungsi ax.scatter() dengan menyertakan 3 buah nilai untuk merepresentasikan koordinat x,y, dan z

```python
z = np.linspace(0,1,100) #paramter pertama: nilai awal, paramter kedua nilai akhri, paramter ketiga banyaknya nilai
x = z * np.sin(20*z)
y = z * np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(x,y,z,color='black')#parameter color adalah parameter optional
ax.set(title='Contoh 3D line plot',
       xlabel = 'sumbu X',
       ylabel = 'sumbu Y',
       zlabel = 'sumbu Z')
plt.show()
```

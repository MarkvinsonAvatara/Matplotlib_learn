# Matplotlib Week 1 Pengenalan Matplotlib

Intro

Matplotlib adalah salah satu module dalam Pemrograman Python yang dapat digunakan untuk memvisualisasikan data dalam bentuk grafik (graph)

Dengan Matplotlib, graphs akandigambarkan pada suatu figure, yang di dalamnya dapat terdiri dari satua tau lebih axes

```python
#import modules

%matplotlib inline 
#magic command yang digunakan untuk memastikan agar semua output yg dihasilkan Matplotlib akan ditampilkan di halam notebook yang sama
#wajib digunakan jika menggunakan versi jupyter notebook yang lama
import matplotlib.pyplot as plt
import numpy as np
```

# Membuat Plotting Sederhana

Membuat sebuah figure yang memiliki sebuah axes; melakukan plotting data pada axes

```python
#sample data prep
x=[1,2,3,4]
y=[4,3,2,1]

flg, ax=plt.subplots() #membuat sebuah figure dan sebuah axes
ax.plot(x,y) #melakukan plotting data pada axes
```

```python
#alternative lain dengan menggunakan method plot() pada pyplot untuk melakukan plotting sederhana
x=[1,2,3,4]
y=[4,3,2,1]

plt.plot(x,y)
```

# Mengenal Object Oriented Style untuk melakukan plotting 

Secara eksplisit membuat figures beserta axes, dan memanggil methods dari keduanya

```python
x=np.linspace(0,3,100) 
#linspace digunakan untuk mengenerate sejumlah deret bilangan yang ditentukan oleh parameternya
#parameter pertama merupakan nilai awal 
#parameter  kedua merupakan nilai akhir
#prameter ketiga berarti banyaknya data yang ingin ditampilkan

x
```

```python
flg, ax=plt.subplots()  #membuat sebuah figure dan sebuah ax

#plotting tiga variant data pada axes 
ax.plot(x,x, label='linear')#pangkat satu
ax.plot(x,x**2, label='quadratic') #**2 pangkat dua
ax.plot(x,x**3, label='cubic') #**3 pangkat tiga

ax.set_xlabel('sumbu x') #menamai sumbu x  pada axes
ax.set_ylabel('y label') ##menamai sumbu y  pada axes
ax.set_title('Simple plot') ##memberi judul  pada axes 
ax.legend()#menampilkan legend
```

# Mengenal pyplot Style untuk melakukan plotting

Mengenal pyplot untuk membuat dan mengelola figures dan axes, serta menggunakan fungsi pada pyplot untuk melakukan plotting

```python
x=np.linspace(0,3,100)
x
```

```python
#plotting tiga variant data pada axes
plt.plot(x,x, label='linear')#pangkat satu
plt.plot(x,x**2, label='quadratic') #**2 pangkat dua
plt.plot(x,x**3, label='cubic') #**3 pangkat tiga

plt.xlabel('sumbu x') #menamai sumbu x  pada axes
plt.ylabel('y label') #menamai sumbu y  pada axes
plt.title('Simple plot') #memberi judul  pada axes
plt.legend()#menampilkan legend

```

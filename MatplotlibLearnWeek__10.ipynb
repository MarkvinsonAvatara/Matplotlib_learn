# Pengaturan  Figure Layouts dengan GridSpec

```python
#import modules

%matplotlib inline 
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
```

```python
#check version
print(matplotlib.__version__)
print(np.__version__)
```

# Simple 2x2 layout dengan subplots

```python
siswa = ['Nats','Nat','Zera','Max']
uas= [65,70,90,80]

fig,axes=plt.subplots(ncols=2,nrows=2, constrained_layout=True)
#constrained_layout digunakan untuk mengatur kerapatan antar plot (False== rapat, True== merapikan)

axes[0][0].bar(siswa,uas)                            #object axes 1 (kiri atas)
axes[0][1].pie(uas,labels=siswa)                     #object axes 2 (kanan atas)
axes[1][0].pie(uas,labels=siswa,explode=[0,0,0,0.4]) #object axes 3 (kiri bawah)
axes[1][1].barh(siswa,uas)                           #object axes 4 (kanan bawah)
plt.show()

#note:
#axes['baris']['kolom']
```

# Simple 2x2 dengan GridSpec

```python
#contoh 1
fig=plt.figure(constrained_layout=True)

gs=gridspec.GridSpec(ncols=2, nrows=2, figure=fig)

ax1=fig.add_subplot(gs[0,0])#object axes 1
ax2=fig.add_subplot(gs[0,1])#object axes 2
ax3=fig.add_subplot(gs[1,0])#object axes 3
ax4=fig.add_subplot(gs[1,1])#object axes 4

ax1.bar(siswa,uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa,explode=[0,0,0,0.4])
ax4.barh(siswa,uas)

plt.show()
```

```python
#contoh 2
fig=plt.figure(constrained_layout=True)

gs=fig.add_gridspec(ncols=2, nrows=2)

ax1=fig.add_subplot(gs[0,0])#object axes 1
ax2=fig.add_subplot(gs[0,1])#object axes 2
ax3=fig.add_subplot(gs[1,0])#object axes 3
ax4=fig.add_subplot(gs[1,1])#object axes 4

ax1.bar(siswa,uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa,explode=[0,0,0,0.4])
ax4.barh(siswa,uas)

plt.show()
```

# Rows Span dan Cols Span dengan GridSpec

```python
fig=plt.figure(constrained_layout=True)

gs=fig.add_gridspec(ncols=2, nrows=2)

ax1=fig.add_subplot(gs[0,:])#object axes 1 dan melakukan perluasan kolom yang tadinya dua menjadi satu kolom
#note: notasi : akan menggunakan seluruh posisi kolom 
#ax2=fig.add_subplot(gs[0,1])#object axes 2
ax3=fig.add_subplot(gs[1,0])#object axes 3
ax4=fig.add_subplot(gs[1,1])#object axes 4

ax1.bar(siswa,uas)
#ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa,explode=[0,0,0,0.4])
ax4.barh(siswa,uas)

plt.show()
```

```python
fig=plt.figure(constrained_layout=True)

gs=fig.add_gridspec(ncols=3, nrows=3)

#object axes 1
ax1=fig.add_subplot(gs[0,:])
#posisi axes: barisnya pada index ke 0 dan kolomnya diperluas sampai seluruh kolom 
ax1.set_title('gs[0, :]')

#object axes 2
ax2=fig.add_subplot(gs[1, :-1])
#posisi axes:  barisnya pada index ke 1  dan kolomnya diperluas sampaidengan index ke -1
ax2.set_title('gs[1, :-1]')
ax2.barh(siswa, uas)

#object 3
ax3=fig.add_subplot(gs[1:, -1])
#posisi axes:  barisnya diperluas sampai dengan index ke terakhir index ke 1 dan kolomnya pada index ke -1
ax3.set_title('gs[1:, -1]')
ax3.bar(siswa, uas)

#object axes 4
ax4=fig.add_subplot(gs[-1,0])
#posisi axes: barisnya pada index ke -1 dan kolomnya pada index ke 0
ax4.set_title('gs[-1,0]')

ax5=fig.add_subplot(gs[-1,-2])
#posisi axes: barisnya pada index ke -1 dan kolomnya pada index ke -2
ax5.set_title('gs[-1,-2]')

plt.show()
```

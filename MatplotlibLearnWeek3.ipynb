# Part 1 Multiple Subplots

Intro Part 1

Multiple Subplots yang akan dipelajari :
   - Simple Line Plot
   - Multiple Subplots dengan OO Style
   - Multiple Subplots dengan pyplot Style

```python
#import modules

%matplotlib inline 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#check pager version
print(matplotlib.__version__)
print(np.__version__)
```

# Simple Line Plot

```python
x= np.arange(0.0,2.0,0.01)
#parameter pertama--> angka awal (nilai ini inclusive)
#parameter kedua--> angka terakhir (nilai ini enklusive)
#parameter ketiga--> interval antar nilai
x
```

```python
s= np.sin(2*np.pi*x) #<--code untuk menghitung hasil operasi sinus unutuk setiap data point yang ada di variable x 
s
```

```python
fig, ax=plt.subplots()#OO style

ax.plot(x,s)
ax.set(xlabel='X label',
      ylabel='nilai sin(x)',
      title='visualisasi nilai sin')

ax.grid()#--> menampilkan grid
plt.show()
```

# Multiple Subplots

```python
x1= np.linspace(0.0,5.0,100)
# parameter pertama: nilai minimum (inclusive)
# parameter kedua : nilai max (inclusive)
# parameter ketiga : banyak nilai yang ingin ditampilkan

x2= np.linspace(0.0,2.0,100)

y1=np.cos(2*np.pi *x1)
y2=np.cos(2*np.pi *x2)
```

```python
x1
```

# Multiple Subplots dengan OO Style

```python
fig, (ax1,ax2)=plt.subplots(2,1)
# parameter pertama di method plt.subplots() --> jumlah baris
# parameter kedua di method plt.subplots() --> jumlah kolom
fig.suptitle('Multiple Subplots')

ax1.plot(x1,y1,'ro-')
ax1.set_ylabel('Nilai $cos(x_1)$')

ax2.plot(x2,y2,'b-')
ax2.set_ylabel('Nilai cos(x2)')
ax2.set_xlabel('Nilai x')

plt.show()
```

# Multiple Subplots dengan pyplot Style

```python
plt.subplot(2,1,1)
#paramter pertama: menenutkan jumlah baris
#paramter kedua: menenutkan jumlah kolom
#paramter ketiga: untuk mengendalikan axis
plt.plot(x1,y1, 'ro-')
plt.title('MUltiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2,1,2)
plt.plot(x2,y2,'g.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```

# Part 2 Histogram

Intro Part 2

Di part ini kita akan mempelajari cara membuat histogram dengan Matplotlib

```python
#import modules

%matplotlib inline 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#check pager version
print(matplotlib.__version__)
print(np.__version__)
```

# Sample Data

```python
mu, sigma=100, 15 #nilai mean dan nilai standard deviation

x=mu + sigma * np.random.randn(10000) #normal distribution
x
```

# Histogram dengan Pyplot Style

```python
plt.hist(x,
        bins=50,
        facecolor='b',
        alpha=0.75)
#parameter wajib pertama: dataset
#parameter wajib kedua: bins (digunakan untuk menuntukan jumlah pengelompokan data)
#alpha digunakan untuk mengatur transparansi (semakin kecil nilai alpha, semakin memudar warnanya)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45,500, r'$\mu=100\ \sigma=15$')
plt.grid()

plt.show()
```

# Histogram dengan Object Oriented Style

```python
fig, ax = plt.subplots()

ax.hist(x,
        bins=50,
        facecolor='g',
        alpha=0.75)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

plt.text(45,500, r'$\mu=100\ \sigma=15$')
plt.grid()

plt.show()
```

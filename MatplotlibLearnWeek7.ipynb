# 3D Line Plot

Cara membuat 3D Line Plot
- membuat objek figure dan objek axes
- menambahkan parameter projection='3d' pada fungsi plt.axes()
- Memanggil fungsi ax.plot() dengan menyertakan 3 buah nilai untuk merepresentasikan koordinat x,y, dan z (nilai x,y,z bebas)

```python
z= np.linspace(0,1,50) #paramter pertama: nilai awal, paramter kedua nilai akhri, paramter ketiga banyaknya nilai
x=z*np.sin(20*z)
y=z*np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot(x,y,z,color='magenta') #parameter color adalah parameter optional
ax.set(title='Contoh 3D line plot',
       xlabel = 'sumbu X',
       ylabel = 'sumbu Y',
       zlabel = 'sumbu Z')
plt.show()
```

# 3D Scatter Plot


Cara membuat 3D Scatter Plot

- membuat objek figure dan objek axes
- menambahkan parameter projection='3d' pada fungsi plt.axes()
- Memanggil fungsi ax.scatter() dengan menyertakan 3 buah nilai untuk merepresentasikan koordinat x,y, dan z (nilai x,y,z bebas)

```python
z=np.linspace(0,1,100) #paramter pertama: nilai awal, paramter kedua nilai akhri, paramter ketiga banyaknya nilai
x=z*np.sin(20*z)
y=z*np.cos(20*z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(x,y,z,color='black')#parameter color adalah parameter optional
ax.set(title='Contoh 3D line plot',
       xlabel = 'sumbu X',
       ylabel = 'sumbu Y',
       zlabel = 'sumbu Z')
plt.show()
```

# Part 2 Contour Plot / Contour Chart

```python
#import modules

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#check version
print(matplotlib.__version__)
print(np.__version__)
```

# Sample Dataset


```python
c=np.linspace(-1,1,100)
x,y=np.meshgrid(c, c)
z=np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

fungsi .meshgrid() digunakan untuk membuat data area atau data surface (membuat data dua dimensi) yang ditampung ke dua variable 

# Contour Plot

- Pada dasarnya Contour Plot adalah visualisasi data yang memvisualisasikan data 3d ke format 2d
- untuk membuat Contour Plot kita hanya perlu memanggil fungsi plt.contour() dengan menyertakan 3 buah nilai yang merepresentasikan nilai x,y, dan z

```python
plt.contour(x, y, z)
plt.colorbar()# untuk menampilkan color barnya

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

# Filled Contour Plot

- FIlled Contour Plot akan memenuhi areanya dengan warna (berbeda dari Contour Plot biasa yang hanya memberikan garis)
- untuk membuat Filled Contour Plot kita hanya perlu memanggil fungsi plt.contourf() dengan menyertakan 3 buah nilai yang merepresentasikan nilai x,y, dan z

```python
plt.contourf(x, y, z)
plt.colorbar()# untuk menampilkan color barnya

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

# Part 3 3D Area Plot / 3D Area Chart 

```python
#import modules

%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

note: agar 3d object bisa kita gerakan (dirotate) kita perlu menggunakan magic command %matplotlib notebook 

```python
#check version
print(matplotlib.__version__)
print(np.__version__)
```

# Sampe Dataset

```python
c=np.linspace(-1,1,100)
x,y=np.meshgrid(c, c)
z=np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

# 3D Wireframe

Cara membuat 3D Wireframe

- membuat objek figure dan objek axes
- menambahkan parameter projection='3d' pada fungsi plt.axes()
- Memanggil fungsi ax.plot_wireframe() dengan menyertakan 3 buah nilai untuk merepresentasikan koordinat x,y, dan z (nilai x,y,z bebas)

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(x,y,z,color='grey')#parameter color adalah parameter optional
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

# 3D Surface Plot

Cara membuat 3D Surface Plot

- membuat objek figure dan objek axes
- menambahkan parameter projection='3d' pada fungsi plt.axes()
- Memanggil fungsi ax.plot_surface() dengan menyertakan 3 buah nilai untuk merepresentasikan koordinat x,y, dan z (nilai x,y,z bebas)


```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,z,color='blue')#parameter color adalah parameter optional
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface')

plt.show()
```

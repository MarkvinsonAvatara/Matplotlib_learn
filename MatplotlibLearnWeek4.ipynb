# Part 1 Bar Plot / Bar Chart

```python
#import modules

%matplotlib inline 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#check version
print(matplotlib.__version__)
print(np.__version__)
```

# Simple Bar Plot 

```python
#data prep
data = [25,45,55,125,225]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori, data) 
#parameter pertama sumbu x
#parameter kedua sumbu y
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

# Pengaturan Grid dan Color

```python
#data prep
data = [25,45,55,125,225]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori,data,color='Blue',alpha=0.75)
plt.grid(linestyle='--',
         linewidth=2,
         axis='y',
         alpha=0.25)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

# Horizontal Bar Plot

```python
#data prep
data = [25,45,55,125,225]
kategori = ['A','B','C','D','E']
```

```python
plt.barh(kategori,data)

plt.grid(linestyle='-',
         linewidth=2,
         axis='x',
         alpha=0.25)

plt.xlabel('jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horizontal Bar Plot 1')

plt.show()
```

```python
plt.barh(kategori[::-1], data[::-1])
#Slicing disini '[::-1]' digunakan untuk membalik urutan list

plt.grid(linestyle='-',
         linewidth=2,
         axis='x',
         alpha=0.25)

plt.xlabel('jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horizontal Bar Plot 2')

plt.show()
```

# Stacked Bar Plot

```python
#data prep
data1 = [25,45,55,75,40]
data2 = [40,35,55,20,10]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori,data1,label='Data 1')
plt.bar(kategori,data2,label='Data 2')


plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha = 0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh stacked bar plot 1')

plt.show()
```

```python
plt.bar(kategori,data1,label='Data 1')
plt.bar(kategori,data2,label='Data 2',bottom = data1)
#botom digunakan untuk mengindikasikan data mana yang ingin dijadikan bagian bawah bar

plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha = 0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh stacked bar plot 2')

plt.show()
```

# Grouped Bar Plot

```python
#data prep
data1 = [25,45,55,75,40]
data2 = [40,35,55,20,10]
kategori = ['A','B','C','D','E']

x = np.arange(len(kategori))
x
```

```python
width=0.5
#untuk mengatur lebar Bar

plt.bar(x-width/2,data1,width,label='Data 1')
plt.bar(x+width/2,data2,width,label='Data 2')

plt.xticks(x,kategori)
#plt.xticks() digunakan untuk mengganti nilai sumbu x dengan nilai kategori

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```

# Part 2 Pie Plot / Pie Chart

```python
#import modules

%matplotlib inline 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#check version
print(matplotlib.__version__)
print(np.__version__)
```

# Sample Pie Plot

```python
#data prep
data = [40,35,55,20,10]
kategori = ['A','B','C','D','E']
```

```python
plt.pie(data, labels=kategori, autopct='%1.1f%%',startangle=190)
#parameter labels optional
#parameter autopct digunakan untuk menampilkan presentase dari setiap pada pie plot (optional)
#parameter startangle digunakan untuk melakukan rotasi pada pie plot

plt.title('Contoh Pie Plot')
plt.show()
```

# Pengaturan Legend dan Color

```python
#data prep
data = [40,35,55,20,10]
kategori = ['A','B','C','D','E']
warna=['Pink','Cyan','#0000FF','yellowgreen','skyblue']
```

```python
plt.pie(data, 
        labels=kategori,
        colors=warna, 
        autopct='%1.1f%%',
        startangle=90)

plt.legend()
plt.title('Contoh Pie Plot')
plt.show()
```

# Exploded Pie Plot

```python
#data prep
data = [40,35,55,20,10]
kategori = ['A','B','C','D','E']
warna=['Purple','Cyan','#0000FF','yellowgreen','skyblue']
explode_var=[0.3,0.5,0.2,0.,0.08]
```

```python
plt.pie(data, 
        labels=kategori,
        colors=warna,
        explode=explode_var, 
        autopct='%1.1f%%',
        startangle=90)
#pada parameter explode, semakin besar nilai explode semakin jauh potongan pie plot yang ditarik

plt.title('Contoh Exploded Pie Plot')
plt.show()
```

# Part 3 Table Plot / Table Chart 

```python
#import modules

%matplotlib inline 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#check version
print(matplotlib.__version__)
print(np.__version__)
```

```python
#Data Prep
data_uas=[
    ['Ryan', 90],
    ['Hawks', 92],
    ['Edward', 85],
    ['William',70],
    ['Aldo',72]
]
```

```python
data_uas
```

# Table Plot dengan Object Oriented Style

```python
fig, ax=plt.subplots()

table=plt.table(cellText=data_uas,
               loc='center')

table.set_fontsize(14)
table.scale(1, 4) 

#scale()digunkanakn untuk mangtur ukuran kolom, baris


ax.axis(False)
#ax.axis(False) digunakan supaya axisnya tidak muncul pada output

plt.show()
```

# Table Plot dengan pyplot style

```python
table= plt.table(cellText=data_uas,
                loc='center')

table.set_fontsize(14)
table.scale(1,4)

ax= plt.gca()
ax.axis(False)

plt.show()
```

# Predefined Styles

```python
plt.style.available
#code diatas digunakan untuk melihat Predefined Style apa saja yang disediakan oleh Matplotlib
```

untuk menggunakan predefind style kita hanya cukup menggunkan method plt.style.use()

```python
plt.style.use('ggplot')
```

```python
x= np.linspace(1,10,15)

plt.plot(x,x,'bs-',label='lin')
plt.plot(x,np.log(x),'ro-',label='log')

plt.legend()
plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title('contoh Bar plot')

plt.show

#note b=blue
#     s=square
#     r=red
#     o=bulat(circle)
```

# Menerapkan Multiple Style

```python
plt.style.use(['seaborn','dark_background'])
```

```python
t=np.linspace(1,10,15)

x= np.linspace(1,10,15)

plt.plot(x,x,'gs-',label='lin')
plt.plot(x,np.log(x),'ro-',label='log')

plt.legend()
plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title('contoh Bar plot')

plt.show
```

# Menerapkan Temporary styling


```python
#contoh 1 (sebelum menerapkan temporary styling)
plt.style.use('ggplot')
plt.plot(np.sin(np.linspace(0, 2*np.pi)),'b-o')
plt.show()
```

```python
#contoh 2
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2*np.pi)),'b-o')
#code diatas adalah contoh cara untuk menerapkan temporary styling
plt.show()
```

```python
#contoh 3 (conoth setelah menearpakan)
plt.plot(np.sin(np.linspace(0, 2*np.pi)),'b-o')
plt.show()
#contoh diatas berbeda dengan contoh dua dikarenakan style yang digunakan itu temporary atau sementara
```
